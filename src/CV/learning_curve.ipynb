```python
import pandas as pd
import matplotlib.pyplot as plt

files = {
    "DenseNet121 (LR=1e-4)": "../../data/cv_history/densenet121_history.csv",
    "DenseNet121 (ReduceLR)": "../../data/cv_history/densenet121_reducelr_history.csv",
    "EfficientNetB0 (LR=1e-4)": "../../data/cv_history/efficientnetb0_history.csv",
    "EfficientNetB0 (LR=1e-5)": "../../data/cv_history/efficientnetb0_v1_history.csv",
    "EfficientNetB0 (ReduceLR)": "../../data/cv_history/efficientnetb0_v2_history.csv"
}

fig, axes = plt.subplots(3, 2, figsize=(15, 12))
axes = axes.flatten()

for i, (label, path) in enumerate(files.items()):
    ax = axes[i]
    df = pd.read_csv(path)
    
    ax.plot(df['accuracy'], label='Training Acc', color='blue', linestyle='--')
    ax.plot(df['val_accuracy'], label='Validation Acc', color='orange', linewidth=2)
    
    ax.set_title(label)
    ax.set_xlabel("Epochs")
    ax.set_ylabel("Accuracy")
    ax.legend()
    ax.grid(True, alpha=0.3)

for j in range(len(files), len(axes)):
    axes[j].axis('off')

plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

file_names = {
    "DenseNet121 (LR=1e-4)": "../../data/cv_history/densenet121_history.csv",
    "DenseNet121 (ReduceLR)": "../../data/cv_history/densenet121_reducelr_history.csv",
    "EfficientNetB0 (LR=1e-4)": "../../data/cv_history/efficientnetb0_history.csv",
    "EfficientNetB0 (LR=1e-5)": "../../data/cv_history/efficientnetb0_v1_history.csv",
    "EfficientNetB0 (ReduceLR)": "../../data/cv_history/efficientnetb0_v2_history.csv"
}

colors = ['#1f77b4', '#ff7f0e', '#2ca02c', '#d62728', '#9467bd']
line_styles = ['-', '--', '-.', ':', '-']

# --- SO SNH ACCURACY ---
plt.figure(figsize=(10, 7))
for i, (label, path) in enumerate(file_names.items()):
    df = pd.read_csv(path)
    plt.plot(df['val_accuracy'], label=label, color=colors[i], linestyle=line_styles[i], linewidth=2)
    plt.scatter(len(df)-1, df['val_accuracy'].iloc[-1], color=colors[i], s=30)

plt.title('Validation Accuracy')
plt.xlabel('Epochs')
plt.ylabel('Accuracy')
plt.legend(loc='lower right')
plt.grid(True, linestyle='--', alpha=0.5)
plt.tight_layout()
plt.show()

# --- SO SNH LOSS ---
plt.figure(figsize=(10, 7))
for i, (label, path) in enumerate(file_names.items()):
    df = pd.read_csv(path)
    plt.plot(df['val_loss'], label=label, color=colors[i], linestyle=line_styles[i], linewidth=2)

plt.title('Validation Loss')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.legend(loc='upper right')
plt.grid(True, linestyle='--', alpha=0.5)
plt.tight_layout()
plt.show()
```
